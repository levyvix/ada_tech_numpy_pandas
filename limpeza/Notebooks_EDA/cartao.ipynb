```python
import pandas as pd
```

```python
df = pd.read_csv(
    "C:\\Users\\levyv\\Desktop\\proj\\Ada_Tech_DS_PY_004\\limpeza\\Data\\Data\\cartao2003-202208032023.csv",
    skiprows=18,
    on_bad_lines="skip",
    encoding="latin1",
    sep=";",
    skipfooter=13,
)
```

```python
df.columns = list(map(lambda x: x.strip(), df.columns))
```

```python
df.isna().sum()
```

```python
def presidente(ano):
    if ano <= 2010:
        return "LULA"
    elif ano <= 2016:
        return "DILMA"
    elif ano <= 2018:
        return "TEMER"
    elif ano <= 2022:
        return "BOLSONARO"
```

```python
df_clean = (
    df.drop(columns=["Unnamed: 8", "Unnamed: 9"])
    .assign(**{"DATA PGTO": pd.to_datetime(df["DATA PGTO"], format="%d/%m/%Y")})
    .assign(
        ANO=lambda df: df["DATA PGTO"].dt.year,
        VALOR_NUM=lambda df: pd.to_numeric(
            df["VALOR"]
            .str.replace(".", "")
            .str.replace(",", ".")
            .str.replace("R$", ""),
            errors="coerce",
        ),
        PRES=lambda df: df["ANO"].apply(presidente),
    )
    .dropna(subset=["VALOR_NUM", "CPF/CNPJ FORNECEDOR"])
    .drop(columns=["VALOR"])
)
```
